#### Let's try to better understand India using the Crime in India Dataset

### 1. Let's find out the major reason people being kidnapped in each and every state

```python
df = pd.read_csv("../input/crime-in-india/39_Specific_purpose_of_kidnapping_and_abduction.csv")
df.head()
```

```python
df.isnull().sum()

df = df.replace(np.nan, 0)
```

```python
df["Sub_Group_Name"].unique()
df.columns
```

```python
abDf = df.groupby("Area_Name")[["K_A_Female_Total", "K_A_Male_Total"]].sum()
abDf["Total"] = abDf["K_A_Female_Total"] + abDf[ "K_A_Male_Total"]
abDf = abDf.sort_values(by = ["Total"], ascending = False)
# abDf = abDf["Total"]
# abDf.sort_values(ascending = False)

sns.barplot(x = abDf.index, y = abDf["Total"]).set(title = "Total no. of kidnapping/abduction in each state")
plt.xticks(rotation = 90)
```

### No of Kidnappings happending each year

```python
# No of Kidnappings happending each year

ywAb = df.groupby("Year")["K_A_Female_Total", "K_A_Male_Total"].sum()
ywAb["Total"] = ywAb["K_A_Female_Total"] + ywAb["K_A_Male_Total"]
ywAb = ywAb.sort_values(by = ["Total"])


sns.barplot(x = ywAb.index, y = ywAb["Total"]).set(title = "Kidnapping/Abduction realated crimes Each year")
```

### Which state has more crime against children and women?

```python
# State which has the most crime against Children
dfCac = pd.read_csv("../input/crime-in-india/crime/crime/03_District_wise_crimes_committed_against_children_2001_2012.csv")

dfCac.head()
```

```python
tt = dfCac.groupby("STATE/UT")["Total"].sum().sort_values(ascending = False)[:10]
print("Most crime commited against Childrens are in these states\n")
tt
```

```python
sns.barplot(x = tt.index, y = tt).set(title = "No. of Crime commited agains Childrens")
```

```python

# Crime against women
dfCaw = pd.read_csv("../input/crime-in-india/crime/crime/42_District_wise_crimes_committed_against_women_2001_2012.csv", index_col="STATE/UT")
dfCaw.head()
```

```python
tt = dfCaw.iloc[: ,3:]
tt.head()
```

```python
tt = tt.groupby("STATE/UT").sum()

tt["Total"] = tt["Kidnapping and Abduction"] 
+ tt["Dowry Deaths"] 
+ tt["Assault on women with intent to outrage her modesty"] 
+ tt["Insult to modesty of Women"] 
+ tt["Cruelty by Husband or his Relatives"] 
+ tt["Importation of Girls"]

rr = tt.sort_values(by = "Total", ascending = False)["Total"][:10]
```

```python
print("There are the 10 states in which crimes against women are the highest")
print(rr)
```

```python
sns.barplot(x = rr.index, y = rr).set(title = "No. of crimes against women")
```

### Which state has more number of complaints against police?
